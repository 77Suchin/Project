## Exercise 1

```python
import cv2
import numpy as np
from PIL import Image, ImageDraw

def find_eye_mouth_boxes(image_path):
   
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 

   
    left_eye_box = (145,180, 240, 300)  
    right_eye_box = (275,180, 390, 300)  
    mouth_box = (165, 330, 350, 430)  

    
    annotated_image = Image.fromarray(image)
    draw = ImageDraw.Draw(annotated_image)

    # Drawing bounding boxes around eyes
    draw.rectangle(left_eye_box, outline="red")
    draw.rectangle(right_eye_box, outline="red")
    

    # Drawing bounding boxes around mouth
    draw.rectangle(mouth_box, outline="blue")

    annotated_image.show()

   
    return left_eye_box, right_eye_box, mouth_box


image_path = "C:/Users/suchi/Downloads/img2.jpg"


left_eye_box, right_eye_box, mouth_box = find_eye_mouth_boxes(image_path)

print("Left eye bounding box:", left_eye_box)
print("Right eye bounding box:", right_eye_box)
print("Mouth bounding box:", mouth_box)

cv2.waitKey(0)  
cv2.destroyAllWindows()
```
